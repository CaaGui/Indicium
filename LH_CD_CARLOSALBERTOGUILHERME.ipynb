```python
import numpy as np
import pandas as pd


df_houses = pd.read_csv('teste_indicium_precificacao.csv')



print('\n\n')
####  ####   ######  #######    
#################################       EDA
####  ####   ######  #######  

print('Relatório - EDA')
print('__________________')
print('\n\n\n')


print('Número de linhas e colunas da tabela: {}'.format(df_houses.shape))
print('\n\n')

print('Amostra de 4 linhas dos dados')
print('\n')
print(
    df_houses
    .head()
)
print('\n\n\n')

print('Resumo das variáveis e seus formatos\n')
print(
    df_houses
    .info()
)

# Alterando tipo de dado da variável ultima_review para o formato de data
df_houses['ultima_review'] = pd.to_datetime(df_houses['ultima_review'])
print('\n\n\n')

print('Resumo de quantas variáveis possuem dados em branco ou ausentes\n')
print(
    (df_houses.isna().sum() / df_houses.shape[0])
    .map('{:.5%}'.format)
)
print('''\n\nOs dados apresentam 4 variáveis com valores ausentes. As variáveis nome e
host_name tem 0.03272% e 0.04295% de dados ausentes, repectivamente. Esse valores
são muito pequenos, permitindo sua remoção. Enquanto as variáveis ultima_review e reviews_por_mes
tem 20,6% de seus valores ausentes, acima do valor limite de 15%. Essas variáveis terão seus valores
ausentes substituídos pelo valor zero, minimizando o impacto gerado.
      ''')

df_houses['ultima_review'] = df_houses['ultima_review'].fillna(0)
df_houses['reviews_por_mes'] = df_houses['reviews_por_mes'].fillna(0)
df_houses = df_houses.dropna()

print('Valores ausentes nas variáveis:')
print('nome ',df_houses['nome'].isna().sum())
print('host_name ', df_houses['host_name'].isna().sum())
print('ultima_review ',df_houses['ultima_review'].isna().sum())
print('reviews_por_mes ', df_houses['reviews_por_mes'].isna().sum())

print('\n\n\n\n\n\n\n')
####  ####   ######  #######    
################################# IDENTIFICANDO CADA VARIÁVEL
####  ####   ######  #######  

print('Identificando as variáveis')
print('________________________')
print('________________________')
print('\n')

print('id')
z = df_houses['id'].nunique()
print(f'\nA variável id possui {z} id diferentes')
print('\n')


print('nome')
z = df_houses['nome'].nunique()
print(f'\nA variável nome possui {z} nomes diferentes')
z  = df_houses['nome'].value_counts()[:25]
print('\n\n25 nomes mais comuns:')
print(
    z
)
print('\n\n\n')



print('host_id')
z = df_houses['host_id'].nunique()
print(f'\nA variável host_id possui {z} id diferentes')
z  = df_houses['host_id'].value_counts()[:25]
print('\n\n25 host_id mais comuns:')
print(
    z
)
print('\n\n\n')



print('host_name')
z = df_houses['host_name'].nunique()
print(f'\nA variável host_name possui {z} nomes diferentes')
z  = df_houses['host_name'].value_counts()[:25]
print('\n\n25 host_name mais comuns:')
print(
    z
)
print('\n\n\n')



print('bairro_group')
z = df_houses['bairro_group'].nunique()
print(f'\nA variável bairro_group possui {z} grupos de bairros diferentes')

z  = df_houses['bairro_group'].value_counts()
print('\n\nQuantidade de imóveis\nlistados por grupo de bairros:')
print(
    z
)
print('\n\n\n') 



print('bairro')
z = df_houses['bairro'].nunique()
print(f'\nA variável bairro possui {z} bairros diferentes')

z = df_houses['bairro'].value_counts()[:10]
print('\n\n Os 10 bairros mais listados')
print(
    z
)
print('\n\n\n')



print('latitude')
z = df_houses['latitude'].nunique()
print(f'\nA variável latitude possui {z} latitudes diferentes')

print('\nInformações sobre a variável latitude:')
print(df_houses['latitude'].describe())
print('\n\n\n')



print('longitude')
z = df_houses['longitude'].nunique()
print(f'\nA variável longitude possui {z} longitudes diferentes')

print('\nInformações sobre a variável longitude:')
print(df_houses['longitude'].describe())
print('\n\n\n')



print('room_type')
z = df_houses['room_type'].nunique()
print(f'\nA variável room_type possui {z} tipos de quartos diferentes:')
print(df_houses['room_type'].unique()[0] + ', ' + 
      df_houses['room_type'].unique()[1] + ', ' + 
      df_houses['room_type'].unique()[2])

z = df_houses['room_type'].value_counts()
print('\n\nQuantidade de imóveis\nlistados por tipos de quartos:')
print(
    z
)
print('\n\n\n')


print('price')
print('\nA variável price é o valor por noite em dólares listado pelo anfitrião')
print('\nInformações sobre a variável price:')
print(
    df_houses['price']
    .describe()
)
print('\n\n\n')



print('minimo_noites')
z = df_houses['minimo_noites'].nunique()
print(f'\nA variável minimo_noites possui {z} mínimo de noites diferentes')

z = df_houses['minimo_noites'].value_counts()[:25]
print(f'\n\nAs 25 quantidades mínimas de noites mais comuns:\n\nNoites  Quantidade')
print(
    z
)
print('\n\n\n')



print('numero_de_reviews')
print('\nA variável numero_de_reviews é o número de comentários dados a cada listagem')
print('\nInformações sobre a variável numero_de_reviews:')
df_houses['numero_de_reviews'].describe()
print('\n\n\n')



print('ultima_review')
z = df_houses['ultima_review']
print('\nA variável ultima_review é a data da última avaliação feita')
print('\nInformações sobre a variável ultima_review:')
print(
    z
    .describe(datetime_is_numeric=True)
)
print('\n\n\n')



print('reviews_por_mes')
print('\nA variável reviews_por_mes é o número de avaliações fornecidas por mês')
print('\nInformações sobre a variável reviews_por_mes:')
print(
    df_houses['reviews_por_mes']
    .describe()
)
print('\nAs 10 melhores avalições da variável reviews_por_mes:')
print(
    df_houses['reviews_por_mes']
    .sort_values(ascending=False)[:10]   
)
print('\n\n\n')



print('calculado_host_listings_count')
print('\nA variável calculado_host_listings_count é a quantidade de listagem por host')
print('\nInformações sobre a variável calculado_host_listings_count:')
print(
    df_houses['calculado_host_listings_count']
    .describe()
)
print('\n\n\n')


print('disponibilidade_365')
print('\nA variável disponibilidade_365 é o número de dias em que o anúncio está disponível para reserva')
print('\nInformações sobre a variável disponibilidade_365:')
print(
    df_houses['disponibilidade_365']
    .describe()
)
print('\n\n\n')





print('\n\n\n')
####  ####   ######  #######    
################################# ANÁLISE DOS DADOS
####  ####   ######  #######    

print('\n\n\nANÁLISE')
print('________________')
print('________________')

print('\nResumo estatístico do reviews por mês para cada grupo de bairro')
z = df_houses.groupby(['bairro_group'])['reviews_por_mes'].describe().applymap("{0:.0f}".format).transpose() 
print(z)

print('\nResumo estatístico da quantidade de reviews por grupo de bairro')
z = df_houses.groupby(['bairro_group'])['numero_de_reviews'].describe().applymap("{0:.0f}".format).transpose() 
print(z)

print('''\n\nAnalisando pela ótica de número de reviews por mês, Manhattan e Brooklyn,
são os mais populares em termos absolutos, 16632 e 16446, respectivamente, seguidos
pelo Queens, 4574, com uma diferença significativa. Manhattan com 58 reviews por mês
é o maior valor para um unico imóvel, seguido pelo Queens com 21 e Brooklyn com 14.

Analisando pela ótica da quantidade de reviews, a característica em termos absolutos
se repete com Manhattan e Brooklyn, 16632 e 16446, respectivamente. Ao avaliar
pela média, temos uma surpresa, sendo Staten Island com maior valor de 31 mas sendo 
o de menor valor absoluto em ambas as perspectivas de análise.
''')

print('''\nA tabela a seguir apresenta a quantidade de vezes o grupo de bairro aparece, 
quantos tipos de quarto diferentes existem, quais desses são os que mais aparecem 
e a frequência com que aparecem.
''')
z = df_houses.groupby(['bairro_group'])['room_type'].describe()
print(z)


print('''\n\nOs grupos de bairros mais interessantes para inventir são
Manhattan, Brooklyn e Queens. Manhattan apresenta um destaque para o tipo
de quarto Entire home/apt, Brooklyn e Queens tem como destaque o Private room.
''')


################################# RESUMO ESTATÍSTICO DE REVIEWS POR MÊS

print('\nResumo estatístico de reviews por mês por grupo de bairro e tipo de quarto')
print(
    df_houses.groupby(['bairro_group','room_type'])['reviews_por_mes']
    .describe()
    .applymap("{0:.0f}".format)
)

print('''\n\nObservando o resumo estatístico de reviews por mês, identifica-se que os 5 com mais
reviews são Manhattan com Entire home/apt, 9967, e Private room, 6309, Brooklyn com Enrtire home/apt,
8164, e Private room, 7992, e o Queens com Private room, 2680. As médias e desvios padrões são prati-
mente iguais, variando entre 1 e 2, e contendo 75% dos valores até 2 ou 3. Por último, Manhattan
Private room com o máximo valor, 58, sendo uma enorme diferença com todos os outros.
''')




################################# RESUMO ESTATÍSTICO DE NÚMERO REVIEWS

print('\nResumo estatístico do número de reviews por grupo de bairro e tipo de quarto')
print(
    df_houses.groupby(['bairro_group','room_type'])['numero_de_reviews']
    .describe()
    .applymap("{0:.0f}".format)
)

print('''\n\nObservando o resumo estatístico de número de reviews, identifica-se uma semelhança, repete-se 
os mesmos locais e tipo de quartos do resumo anterior. Assim sendo os 5 com mais reviews são Manhattan com 
Entire home/apt, 13199, e Private room, 7982, Brooklyn com Enrtire home/apt, 9559, e Private room, 101310,
e o Queens com Private room, 3372. Destaque para as médias, sem relação alguma com os 5 mesmos citados, sendo 
as 3 maiores em Staten Island com os tipos de quartos Entire home/apt, 33, Private room, 30, e no Bronx Entire
home/apt, 31. Sendo esses, entre os com menores presença(frequência) e menores máximas.
''')





################################# RESUMO ESTATÍSTICO DO MÍNIMO DE NOITES



print('\n\nResumo estatístico do mínimo de noites por grupo de bairro e tipo de quarto')
print(
    df_houses.groupby(['bairro_group','room_type'])['minimo_noites']
    .describe()
    .iloc[:,1:]
    .applymap("{0:.0f}".format)
    #.transpose()
)



################################# RESUMO ESTATÍSTICO DO PREÇO

print('\n\nResumo estatístico do preço por grupo de bairro e tipo de quarto')
print(
    df_houses.groupby(['bairro_group','room_type'])['price']
    .describe()
    .iloc[:,1:]
    .applymap("${0:.0f}".format)
    #.transpose()
)


print('''\n\nCom os 2 últimos resumos sobre mínimo de noites e preço, o destaque está para
Manhattan Entire home/apt, com a maior média e o segundo maior desvio padrão tanto em noites 
quanto em preço. O maior número de noites, 1250, e preço máximo de 10 mil dólares, apresenta 
a possibilidade de uma margem de lucro alta. Assim sendo a melhor opção como investimento.
''')




################################# MATRIZ DE CORRELAÇÃO DO PREÇO


print('\nMatriz correlação do preço')
z = df_houses[['price','minimo_noites','disponibilidade_365']].corr()
print(z)

print('''\nAs variáveis que representam o mínimo de noites e a disponibilidade ao 
longo do ano não inteferem nos preços, apresentam uma correlação muito proxima de 0.
''')



################################# BUSCA DAS PALAVRA MAIS PRESENTES

print('\n\nAs 20 palavras mais utilizadas no nome do imóvel/local de alto valor')
filtro = (df_houses['price'] >= 175)
words = []
z = df_houses.loc[filtro,['nome']]
for i in range(len(z)):
    for word in str(z.iloc[i,0]).split():
        words.append(word)
    
print(pd.Series(words).value_counts()[:20])

print('''\nLevando em consideração a definição de alto valor sendo locais onde o preço 
é igual ou superior a 175 dólares, temos as palavras mais comuns:\n
    - o tipo de quarto ou de locação
    - o tipo de apartamento/estabelecimento
    - a zona em que se encontra
    - nome do bairro, no caso os bairros de Manhattan e Williamsburg
    - as palavras: luxury e spacious
''')




print('\n\n\n\n\n\n\n\n\n\n')
####  ####   ######  ####### 
################################# INFORMAÇÕES ADICIONAIS
####  ####   ######  ####### 
 
print('''                      Informações adicionais sobre os grupos de bairro.    
                          ______________________________________
                                          ______
''')

grupos_de_bairros = ['Manhattan','Brooklyn','Queens', 'Bronx', 'Staten Island']

for grupo_de_bairro in grupos_de_bairros:
    
    bairro_group = grupo_de_bairro    
    print(f'\n\n\n{bairro_group}')
    print('________________')
    print('________________')

    filtro = (df_houses['bairro_group'] == bairro_group)
    z = df_houses.loc[filtro,:]


    bairros = z['bairro'].nunique()
    print(f'\nO grupo de bairro {bairro_group} possui {bairros} bairros')
    
    
    ################################# RESUMO ESTATÍSTICO DO PREÇO

    print(f'\n\nResumo estatístico do preço {bairro_group} por bairro')
    print('\n')
    print(
        z.groupby(['bairro'])['price']
        .describe()
        .sort_values(by='count',ascending=False)
        .iloc[:,1:]
        .applymap("${0:.0f}".format)
    )

    ################################# RESUMO ESTATÍSTICO DO MÍNIMO DE NOITES

    print(f'\n\nResumo estatístico do mínimo de noites em {bairro_group} por bairro - 10 maiores')
    print('\n')
    print(
        z.groupby(['bairro'])['minimo_noites']
        .describe()
        .sort_values(by='count',ascending=False)
        .iloc[:,1:]
        .applymap("{0:.2f}".format)
    )



    ################################# RESUMO ESTATÍSTICO DO NÚMERO DE REVIEWS

    print(f'\n\nResumo estatístico do número de reviews por bairro de {bairro_group}')
    print('\n')
    print(
        z.groupby(['bairro'])['numero_de_reviews']
        .describe()
        .sort_values(by='count',ascending=False)
        .iloc[:,1:]
        .applymap("{0:.2f}".format)
    )



    ################################# RESUMO ESTATÍSTICO DE REVIEWS POR MÊS

    print(f'\n\nResumo estatístico de reviews por mês por bairro de {bairro_group}')
    print('\n')
    print(
        z.groupby(['bairro'])['reviews_por_mes']
        .describe()
        .sort_values(by='count',ascending=False)
        .applymap("{0:.2f}".format)
    )



    ################################# RESUMO HOST NAME

    print(f'\n\nTop 10 nomes de host em {bairro_group}')
    print('\nHost Name     Imóveis')
    print(
        z['host_name']
        .value_counts()[:10]
    )
```
